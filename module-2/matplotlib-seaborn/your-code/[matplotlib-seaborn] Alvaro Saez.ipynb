#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
#PANDAS
#no hara falta pivot table pero as practico
# MEDIAN STEPS BY MONTH_ENCODED
fitbit_pt_sbm = fitbit_db.pivot_table(index =["Months_encoded"],values =["Steps"])
fitbit_pt_sbm["Steps"] = fitbit_pt_sbm["Steps"].round(0).astype(int)
# your code here
fitbit_pt_sbm.plot.bar(y="Steps",figsize=(12, 4))
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_pt_sbw = fitbit_db.pivot_table(index =["Work_or_Weekend"],values =["Steps"])
fitbit_pt_sbw["Steps"] = fitbit_pt_sbw["Steps"].round(0).astype(int)
fitbit_pt_sbw.index = np.where(fitbit_pt_sbw.index == 0, "weekend", "week")
fitbit_pt_sbw.plot.barh(y="Steps",figsize=(12, 4))
plt.show()
```

```python
#Matplotlib

# MEDIAN STEPS BY MONTH_ENCODED
fitbit_pt_sbm = fitbit_db.pivot_table(index =["Months_encoded"],values =["Steps"])
fitbit_pt_sbm["Steps"] = fitbit_pt_sbm["Steps"].round(0).astype(int)
# your code here
fig_ch2_sbm, ax_ch2_sbm =  plt.subplots(nrows=1, ncols=1,figsize=(12, 4))
ax_ch2_sbm.bar(fitbit_pt_sbm.index.tolist(),fitbit_pt_sbm["Steps"].tolist(), label='steps')
ax_ch2_sbm.legend()
ax_ch2_sbm.set_xlabel("MONTHS")
ax_ch2_sbm.set_ylabel("MEDIAN STEPS")
ax_ch2_sbm.set_title("MEDIAN STEPS BY MONTH_ENCODED")

plt.savefig("../figures/median_steps_by_month.png")
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_pt_sbw = fitbit_db.pivot_table(index =["Work_or_Weekend"],values =["Steps"])
fitbit_pt_sbw["Steps"] = fitbit_pt_sbw["Steps"].round(0).astype(int)
fitbit_pt_sbw.index = np.where(fitbit_pt_sbw.index == 0, "weekend", "week")

fig_ch2_sbw, ax_ch2_sbw =  plt.subplots(nrows=1, ncols=1,figsize=(12, 4))
ax_ch2_sbw.barh(fitbit_pt_sbw.index.tolist(),fitbit_pt_sbw["Steps"].tolist(), label='steps')
ax_ch2_sbw.legend()
ax_ch2_sbw.set_xlabel("MEDIAN STEPS")
ax_ch2_sbw.set_title("MEDIAN STEPS BY WORK_OR_WEEKEND")
ax_ch2_sbw.text(fitbit_pt_sbw["Steps"][0],"weekend",str(fitbit_pt_sbw["Steps"][0]), color="black")
ax_ch2_sbw.text(fitbit_pt_sbw["Steps"][1],"week",str(fitbit_pt_sbw["Steps"][1]), color="black")

plt.savefig("../figures/week_or_week_end.png")
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here
fig_titanic_age_bins, ax_titanic_age_bins =  plt.subplots(nrows=1, ncols=1,figsize=(12, 6))
ax_titanic_age_bins.hist(titanic_db["Age"].tolist(), bins = [0,20,50,80])
ax_titanic_age_bins.set_xlabel("Age")
ax_titanic_age_bins.set_ylabel("N of Passengers")
ax_titanic_age_bins.set_title("Age Histogram")
plt.style.use('seaborn-muted')
plt.show()
```
